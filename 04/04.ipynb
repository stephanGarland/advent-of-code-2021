```python
from pprint import pprint
import sys

import numpy as np

sys.path.append("..")
from classes.utilities import Utilities
```

```python
ip, _ = Utilities.get_input()
```

```python
def rle(inarray):
        ia = np.asarray(inarray)                # force numpy
        n = len(ia)
        if n == 0: 
            return (None, None, None)
        else:
            y = ia[1:] != ia[:-1]               # pairwise unequal (string safe)
            i = np.append(np.where(y), n - 1)   # must include last element posi
            z = np.diff(np.append(-1, i)).tolist()       # run lengths
            p = np.cumsum(np.append(0, z))[:-1].tolist() # positions
            return (z, p)
```

```python
draw_order = [int(x) for x in ip[0].split(',')]
boards = []
for i in range(1, len(ip[1:])-1, 5):
    # Have to first strip out whitespace, then split to get numbers as elements
    line = [x.strip().split() for x in ip[i:i+5]]
    int_line = [list(map(int, x)) for x in line]
    boards.append(int_line)
boards_np = np.array(boards)
```

```python
def run():
    numbers = []
    for num in draw_order:
        print(f"Drawing {num}")
        # Return the indices of numbers matching the number called
        match_np = np.nonzero(boards_np == num)
        # Turn those into board/row/col
        board_match, row_match, col_match = list(zip(*match_np))[0]
        # Complement the matches to mark them
        boards_np[board_match][row_match][col_match] = ~num
        # Return the indices of any negative numbers we've marked
        board_flip, row_flip, col_flip = list(zip(*np.argwhere(boards_np < 0)))
        for board in board_flip:
            print(f"Using {board}")
            # Get RLE for the row of a given board
            row_run, row_pos = rle(row_flip)
            for i, run_len in enumerate(row_run):
                if run_len == 5:
                    remaining = boards_np[row_flip[i]]
                    remaining_sum = np.where(remaining>0, remaining, 0).sum()
                    return board_match, boards[board_match], remaining_sum, num
            # Get RLE for the col of a given board
            run_col, pos_col = rle(col_flip)
            for j, run_len in enumerate(run_col):
                if run_len == 5:
                    remaining = boards_np[col_flip[i]]
                    remaining_sum = np.where(remaining>0, remaining, 0).sum()
                    return boards_np[board_match], remaining_sum, num
```
